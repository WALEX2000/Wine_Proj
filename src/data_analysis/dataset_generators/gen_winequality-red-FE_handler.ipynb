# Data Handler Generator for winequality-red-FE.csv Raw Data File

```python
from qml_custom.data_handler import DataHandler
```

```python
dataPath = '../../../data/winequality-red-FE.csv'
targetVarName = 'quality'
def rmse(*args, **kwargs):
    from sklearn.metrics import mean_squared_error
    return mean_squared_error(*args, **kwargs, squared=False)

def f1(*args, **kwargs):
    from sklearn.metrics import f1_score
    return f1_score(*args, **kwargs, average="macro")

def score(*args, **kwargs):
    from sklearn.metrics import mean_squared_error, mean_absolute_error, f1_score, accuracy_score
    rmse = mean_squared_error(*args, **kwargs, squared=False)
    mae = mean_absolute_error(*args, **kwargs)
    f1 = f1_score(*args, **kwargs, average=None)
    f1_weighted = f1_score(*args, **kwargs, average="weighted")
    acc = accuracy_score(*args, **kwargs)
    print(f"RMSE: {rmse}")
    print(f"MAE: {mae}")
    print(f"F1: {f1}")
    print(f"F1 Weighted: {f1_weighted}")
    print(f"Acc: {acc}")

    return mae

scoringFunction = score
dataHandler = DataHandler(targetVarName, scoringFunction, dataPath=dataPath, maximize=False)
dataHandler.getDataframe().head()
```

```python
# Do NOT edit the save path
savePath = '../../../data/data_conf/winequality-red-FE.csv.handler'
dataHandler.save(savePath)
```
